```python
import dotenv
dotenv.load_dotenv()
```

```python
import anthropic

client = anthropic.Anthropic()
model_id = "claude-sonnet-4-20250514"

messages = [{
    "role": "user",
    "content": "Hello, Claude",
}]

message = client.messages.create(
    model=model_id,
    max_tokens=1000,
    temperature=0,
    messages=messages
)

print(message.content[0].text)
```

Make sure you create a .env file like this 'ANTHROPIC_API_KEY=""'
