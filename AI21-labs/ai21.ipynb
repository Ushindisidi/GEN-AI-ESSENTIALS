```python
import os
import dotenv
dotenv.load_dotenv()
from ai21 import AI21Client
from ai21.models.chat import ResponseFormat
from ai21.models.chat import UserMessage

#new imports
from ai21 import AI21Client
from ai21.models.chat import ChatMessage
```

```python


# Make sure your AI21_API_KEY environment variable is set
# For example: os.environ["AI21_API_KEY"] = "YOUR_API_KEY_HERE"
client = AI21Client(api_key=os.environ.get("AI21_API_KEY"))

# Create chat completion
response = client.chat.completions.create(
    model="jamba-large", 
    messages=[
        ChatMessage(role="user", content="Tell me something I don't know. Limit the response to 30 words maximum.")
    ],
    max_tokens=50,
    temperature=0.4,
    top_p=1.0,
)

# Print the response
print(response.choices[0].message.content)
```

```python
import os
from ai21 import AI21Client
from ai21.models.chat import ChatMessage

client = AI21Client(api_key=os.environ.get("AI21_API_KEY"))

# Create the chat completion with a system message
response = client.chat.completions.create(
    model="jamba-large",
    messages=[
        ChatMessage(role="system", content="You are a helpful and concise assistant."),
        ChatMessage(role="user", content="Explain the concept of photosynthesis in one sentence."),
    ],
    max_tokens=100,
    temperature=0.7,
    top_p=1.0,
)

# Print the response
print(response.choices[0].message.content)
```
